```python
class customNeuralNetwork:

    def __init__(self):

        self.w1 = 1
        self.w2 = 1
        self.bias = 0

    def sigmoid_np(self , X):
        
       return 1 / (1 + np.exp(-X))

    def log_loss(self , y_true , y_predicted ):
       
       epsilon = 1e-15

       y_predicted = np.clip(y_predicted, epsilon, 1 - epsilon)


       y_predicted = np.array(y_predicted)

       return -1 * np.mean(y_true * np.log(y_predicted) + (1 - y_true) * np.log(1 - y_predicted))
       

    def gradient_descent(self , age , affordibility , y_true , epochs ):

        learning_rate = 0.5

        n = len(age)

        for i in range(epochs):

            weighted_sum = self.w1 * age + self.w2 * affordibility + self.bias 

            y_predicted = self.sigmoid_np(weighted_sum)
   
            loss = self.log_loss( y_true , y_predicted)
            
            w1_d = 1/n * np.dot(np.transpose(age) , (y_predicted - y_true))

            w2_d = 1/n * np.dot(np.transpose(affordibility) , (y_predicted - y_true))

            bias_d = np.mean(y_predicted - y_true)
            
            self.w1 -= learning_rate * w1_d
            
            self.w2 -= learning_rate * w2_d
            
            self.bias -= learning_rate * bias_d
            
            if i % 50 == 0:
                
                print(f"Epoch : {i} , w1 : {self.w1} , w2 : {self.w2} , loss : {loss} , bias : {self.bias}")

            if loss <= 0.4719:

                print(f"Epoch : {i} , w1 : {self.w1} , w2 : {self.w2} , loss : {loss} , bias : {self.bias}")

                break

        return self.w1 , self.w2 , self.bias

                
                


        
        

      
    

    
         
       
       
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("C:/Users/Abhi/Downloads/Gradient_Descent_From_Scratch/insurance_data (1).csv")
```

```python
df.head()
```

```python
X_train , X_test , y_train , y_test = train_test_split(df[['age','affordibility']] , df['bought_insurance'] , test_size = 0.2 , random_state = 25)
```

```python
X_train_scaled = X_train.copy()
X_train_scaled['age'] = X_train['age'] /100

X_test_scaled = X_test.copy()
X_test_scaled['age'] = X_test['age'] / 100
```

```python
X_test_scaled.head()
```

```python
from tensorflow import keras
```

```python
model = keras.Sequential(
    [keras.layers.Dense(1 , input_shape = (2,) , activation = 'sigmoid')])

model.compile(optimizer= 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])
```

```python
model.fit(X_train_scaled , y_train , epochs = 1000)
```

```python
model.predict(X_test_scaled)
```

```python
model.evaluate(X_test_scaled , y_test)
```
